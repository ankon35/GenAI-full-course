Dynamic Prompts

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate
llm = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0)
```

```python
prompts = ChatPromptTemplate.from_messages([
    {"role": "system" , "content" : "You are a {developer_type} developer"},
    {"role": "user" , "content" : "{query}"}
])


# final_prompts = prompts.format_prompt(developer_type="python", query="How to sort an array?")
# res = llm.invoke(final_prompts)
# print(res.content)
```

```python
chains = prompts | llm

res = chains.invoke({"developer_type": "java", "query": "How to create a class in Java?"})
print(res.content)
```
